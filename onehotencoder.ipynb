```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies= pd.get_dummies(df.town)
dummies
```

```python
mearged = pd.concat([df,dummies],axis= 'columns')
mearged
```

```python
final=mearged.drop(['town','west windsor'],axis='columns')
final
```

```python
model=linear_model.LinearRegression()
```

```python
#price is dependent variable drop it
x=final.drop(['price'],axis='columns')
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[2800,1,0]])
```

```python
user_input=input('write Parameters')
value=[int(i)for i in user_input.split(',')]
predict=model.predict([value])
print('priducted : ',predict[0])
```

```python
model.score(x,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.town=le.fit_transform(dfle.town)
df
```

```python
x=dfle[['town','area']].values
x
```

```python
y=dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
columnTransformer = ColumnTransformer(
    transformers=[
        ('town', OneHotEncoder(drop='first', sparse_output=False), [0])  # use sparse_output=False here
    ],
    remainder='passthrough'  # keep other columns like 'price'
)
```

```python
# Transform
x = columnTransformer.fit_transform(x)
x
```

```python
model.fit(x,y)
```

```python
model.predict([[1,0,2800]])
```
